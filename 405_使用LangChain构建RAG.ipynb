# simple_rag_v3.py

```python
# å¯¼å…¥ itemgetterï¼Œç”¨äºä»æ•°æ®ç»“æ„ä¸­æå–ç‰¹å®šå­—æ®µ
from operator import itemgetter
# å¯¼å…¥ Pathï¼Œç”¨äºå¤„ç†æ–‡ä»¶è·¯å¾„
from pathlib import Path
# å¯¼å…¥æç¤ºè¯æ¨¡æ¿ç±»
from langchain_core.prompts import PromptTemplate
# å¯¼å…¥è¾“å‡ºè§£æå™¨ï¼Œå°†æ¨¡å‹è¾“å‡ºè§£æä¸ºå­—ç¬¦ä¸²
from langchain_core.output_parsers import StrOutputParser
# å¯¼å…¥å¯ç»„åˆçš„è¿è¡Œå•å…ƒ
from langchain_core.runnables import RunnablePassthrough, RunnableParallel
# å¯¼å…¥æ–‡ä»¶å¤¹åŠ è½½å™¨å’Œæ–‡æœ¬æ–‡ä»¶åŠ è½½å™¨
from langchain_community.document_loaders import DirectoryLoader, TextLoader
# å¯¼å…¥æ–‡æœ¬åˆ‡åˆ†å™¨ï¼Œç”¨äºå¯¹æ–‡æ¡£è¿›è¡Œåˆ†å—
from langchain_text_splitters import RecursiveCharacterTextSplitter, CharacterTextSplitter
# å¯¼å…¥ HuggingFace å‘é‡åµŒå…¥æ¨¡å‹
from langchain_huggingface import HuggingFaceEmbeddings
# å¯¼å…¥ OpenAI çš„èŠå¤©æ¨¡å‹ï¼ˆå…¼å®¹ DeepSeek æ¥å£ï¼‰
from langchain_openai import ChatOpenAI
# å¯¼å…¥ Chroma å‘é‡æ•°æ®åº“
from langchain_chroma import Chroma
# å†æ¬¡å¯¼å…¥æç¤ºè¯æ¨¡æ¿ç±»ï¼ˆå¯çœç•¥ï¼Œå·²åœ¨å‰é¢å¯¼å…¥ï¼‰
from langchain.prompts import PromptTemplate
# å¯¼å…¥ os æ¨¡å—ï¼Œç”¨äºè¯»å–ç³»ç»Ÿç¯å¢ƒå˜é‡
import os

# ä»ç¯å¢ƒå˜é‡ä¸­è·å–åä¸º DEEPSEEK_API_KEY çš„å¯†é’¥ï¼ˆä½ éœ€æå‰è®¾ç½®è¿™ä¸ªç¯å¢ƒå˜é‡ï¼‰
api_key = os.environ['DEEPSEEK_API_KEY']

# å®ä¾‹åŒ–å¯¹è¯è¯­è¨€æ¨¡å‹ï¼ˆå¤§è¯­è¨€æ¨¡å‹ï¼‰ï¼Œä½¿ç”¨ deepseek-chatï¼Œä¼ å…¥ API å¯†é’¥
llm = ChatOpenAI(model='deepseek-chat', base_url='https://api.deepseek.com', api_key=api_key)

# å®ä¾‹åŒ–å‘é‡åµŒå…¥æ¨¡å‹ï¼Œç”¨äºå°†æ–‡æœ¬è½¬æ¢ä¸ºå‘é‡ï¼Œè·¯å¾„å¿…é¡»ä¸ºæœ¬åœ° HuggingFace æ¨¡å‹è·¯å¾„
embedding_model = HuggingFaceEmbeddings(model=r'model_dir\BAAI\bge-large-zh-v1___5')

# æŒ‡å®šåŒ…å«çŸ¥è¯†æ–‡æœ¬çš„æœ¬åœ°æ–‡ä»¶å¤¹è·¯å¾„
file_dir = Path('my_knowledge')

# æ–‡æœ¬åˆ‡åˆ†å™¨ï¼šå°†æ–‡æ¡£æŒ‰ 500 å­—ä¸ºä¸€å—è¿›è¡Œåˆ‡åˆ†ï¼Œé‡å éƒ¨åˆ†ä¸º 100 å­—
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)

# åˆ›å»º Chroma å‘é‡æ•°æ®åº“å¯¹è±¡ï¼ŒæŒ‡å®šåµŒå…¥æ¨¡å‹å’Œä¿å­˜ä½ç½®
vector_store = Chroma(embedding_function=embedding_model, persist_directory='./chroma_v3')

# è®¾ç½®æ£€ç´¢å™¨ï¼Œç”¨äºä»å‘é‡æ•°æ®åº“ä¸­æŒ‰ç›¸ä¼¼åº¦å–å‡ºå‰ k ä¸ªæ–‡æ¡£ï¼ˆé»˜è®¤ 5 ä¸ªï¼‰
retriever = vector_store.as_retriever(search_kwargs={'k': 5})

# åˆ›å»ºæç¤ºè¯æ¨¡æ¿ï¼š{context} å’Œ {question} æ˜¯å¯æ›¿æ¢çš„å ä½ç¬¦
prompt_template = PromptTemplate.from_template("""
ä½ æ˜¯ä¸€ä¸ªä¸¥è°¨çš„RAGåŠ©æ‰‹ã€‚
è¯·æ ¹æ®ä»¥ä¸‹æä¾›çš„ä¸Šä¸‹æ–‡ä¿¡æ¯æ¥å›ç­”é—®é¢˜ã€‚
å¦‚æœä¸Šä¸‹æ–‡ä¿¡æ¯ä¸è¶³ä»¥å›ç­”é—®é¢˜ï¼Œè¯·ç›´æ¥è¯´â€œæ ¹æ®æä¾›çš„ä¿¡æ¯æ— æ³•å›ç­”â€
å¦‚æœå›ç­”æ—¶ä½¿ç”¨äº†ä¸Šä¸‹æ–‡ä¸­çš„ä¿¡æ¯ï¼Œåœ¨å›ç­”åè¾“å‡ºä½¿ç”¨äº†å“ªäº›ä¸Šä¸‹æ–‡ï¼Œ
ä¸Šä¸‹æ–‡ä¿¡æ¯:
{context}
-------------
é—®é¢˜ï¼š{question}""")

# åˆ›å»ºä¸€ä¸ªé“¾ï¼ˆChainï¼‰ï¼ŒåŒ…å«ä»¥ä¸‹æ­¥éª¤ï¼š
# 1. æ¥æ”¶ç”¨æˆ·è¾“å…¥çš„ 'question'ï¼›
# 2. å°† question ä¼ å…¥ retrieverï¼Œè·å–ç›¸å…³ contextï¼›
# 3. ä½¿ç”¨ prompt_template æ„é€ å®Œæ•´æç¤ºè¯ï¼›
# 4. ä½¿ç”¨ LLM ç”Ÿæˆå›ç­”ï¼›
# 5. ä½¿ç”¨ StrOutputParser è§£æè¾“å‡ºä¸ºå­—ç¬¦ä¸²ã€‚
chain = {
    'question': RunnablePassthrough()  # åŸæ ·ä¼ é€’ question
} | RunnablePassthrough.assign(        # å¢åŠ  context å­—æ®µï¼ˆä» retriever ä¸­æ£€ç´¢ question ç›¸å…³å†…å®¹ï¼‰
    context=itemgetter('question') | retriever
) | prompt_template | llm | StrOutputParser()  # æ„é€  promptã€è°ƒç”¨æ¨¡å‹ã€è§£æè¾“å‡º

# æ‰§è¡Œä¸»ç¨‹åº
if __name__ == '__main__':
    # ä½¿ç”¨ DirectoryLoader åŠ è½½ my_knowledge ç›®å½•ä¸­çš„æ‰€æœ‰æ–‡æœ¬æ–‡ä»¶
    # TextLoader ç”¨äºä»¥ UTF-8 ç¼–ç æ–¹å¼è¯»å–æ¯ä¸ªæ–‡ä»¶
    docs = DirectoryLoader(str(file_dir),
        loader_cls=lambda path: TextLoader(path, encoding='utf-8')
    ).load()

    # å°†åŠ è½½çš„æ–‡æ¡£è¿›è¡Œåˆ‡åˆ†ï¼Œåˆ†æˆå¤šä¸ªå°æ®µ
    docs = text_splitter.split_documents(docs)

    # å°†åˆ‡åˆ†å¥½çš„æ–‡æ¡£æ·»åŠ åˆ°å‘é‡æ•°æ®åº“ä¸­è¿›è¡Œå­˜å‚¨
    vector_store.add_documents(docs)

    # æµ‹è¯•é“¾æ¡ï¼šè¾“å…¥ä¸€ä¸ªé—®é¢˜â€œèƒ½é£å¤šä¹…ï¼Ÿâ€ï¼Œå¹¶æ‰“å°æ¨¡å‹çš„å›ç­”
    print(chain.invoke('èƒ½é£å¤šä¹…ï¼Ÿ'))
```
