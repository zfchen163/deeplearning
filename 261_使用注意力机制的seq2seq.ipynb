```python
import torch
from torch import nn
from d2l import torch as d2l
import os
```

```python
# å¸¦æœ‰æ³¨æ„åŠ›æœºåˆ¶çš„è§£ç å™¨åŸºæœ¬æ¥å£
class AttentionDecoder(d2l.Decoder):
    """å¸¦æœ‰æ³¨æ„åŠ›æœºåˆ¶çš„è§£ç å™¨åŸºæœ¬æ¥å£"""
    def __init__(self, **kwargs):
        super(AttentionDecoder, self).__init__(**kwargs)
        
    @property
    def attention_weight(self):
        raise NotImplementedError
```

```python
# å®ç°å¸¦æœ‰Bahdanauæ³¨æ„åŠ›çš„å¾ªç¯ç¥ç»ç½‘ç»œè§£ç å™¨
class Seq2SeqAttentionDecoder(AttentionDecoder):
    def __init__(self, vocab_size, embed_size, num_hiddens, num_layers, dropout=0, **kwargs):
        # è°ƒç”¨çˆ¶ç±»AttentionDecoderçš„æ„é€ å‡½æ•°è¿›è¡Œåˆå§‹åŒ–
        super(Seq2SeqAttentionDecoder, self).__init__(**kwargs)
        # åˆ›å»ºä¸€ä¸ªåŠ æ€§æ³¨æ„åŠ›æœºåˆ¶çš„å®ä¾‹ï¼Œç”¨äºè®¡ç®—æ³¨æ„åŠ›æƒé‡
        self.attention = d2l.AdditiveAttention(num_hiddens, num_hiddens, num_hiddens, dropout)  
        # åˆ›å»ºä¸€ä¸ªåµŒå…¥å±‚ï¼Œç”¨äºå°†è¾“å…¥çš„æ•´æ•°åºåˆ—è¿›è¡ŒåµŒå…¥è¡¨ç¤º
        self.embedding = nn.Embedding(vocab_size, embed_size)
        # åˆ›å»ºä¸€ä¸ªGRUå±‚ï¼Œç”¨äºå®ç°å¾ªç¯ç¥ç»ç½‘ç»œçš„è®¡ç®—
        self.rnn = nn.GRU(embed_size + num_hiddens, num_hiddens, num_layers, dropout=dropout)  
        # åˆ›å»ºä¸€ä¸ªçº¿æ€§å±‚ï¼Œå°†éšè—çŠ¶æ€æ˜ å°„åˆ°è¯æ±‡è¡¨å¤§å°çš„è¾“å‡º
        self.dense = nn.Linear(num_hiddens, vocab_size)
        
    def init_state(self, enc_outputs, enc_valid_lens, *args):
        # å°†ç¼–ç å™¨çš„è¾“å‡ºè§£åŒ…ä¸ºoutputså’Œhidden_state
        outputs, hidden_state = enc_outputs
        # å¯¹outputsè¿›è¡Œç»´åº¦å˜æ¢ï¼Œå°†batchç»´å’Œæ—¶é—´æ­¥ç»´äº¤æ¢ï¼Œä¿æŒä¸è§£ç å™¨è¾“å…¥çš„ä¸€è‡´æ€§
        # è¿”å›åˆå§‹åŒ–çš„è§£ç å™¨éšè—çŠ¶æ€
        return (outputs.permute(1, 0, 2), hidden_state, enc_valid_lens)
    
    def forward(self, X, state):
        # è§£æè¾“å…¥çš„çŠ¶æ€ä¿¡æ¯ï¼ŒåŒ…æ‹¬ç¼–ç å™¨çš„è¾“å‡ºã€éšè—çŠ¶æ€å’Œæœ‰æ•ˆé•¿åº¦
        enc_outputs, hidden_state, enc_valid_lens = state
        # å¯¹è¾“å…¥åºåˆ—è¿›è¡ŒåµŒå…¥è¡¨ç¤ºï¼Œå¹¶è¿›è¡Œç»´åº¦å˜æ¢ï¼Œå°†batchç»´å’Œæ—¶é—´æ­¥ç»´äº¤æ¢
        X = self.embedding(X).permute(1, 0, 2)
        # åˆå§‹åŒ–è¾“å‡ºåˆ—è¡¨å’Œæ³¨æ„åŠ›æƒé‡åˆ—è¡¨
        outputs, self._attention_weights = [], []
        for x in X:
            # è·å–å½“å‰æ—¶é—´æ­¥çš„æŸ¥è¯¢å‘é‡ï¼Œå°†éšè—çŠ¶æ€çš„æœ€åä¸€ä¸ªæ—¶é—´æ­¥çš„ç‰¹å¾è¿›è¡Œç»´åº¦æ‰©å±•
            query = torch.unsqueeze(hidden_state[-1], dim=1)
            # è®¡ç®—æ³¨æ„åŠ›ä¸Šä¸‹æ–‡å‘é‡ï¼Œé€šè¿‡æ³¨æ„åŠ›æœºåˆ¶å¯¹ç¼–ç å™¨çš„è¾“å‡ºè¿›è¡ŒåŠ æƒæ±‚å’Œ
            context = self.attention(query, enc_outputs, enc_outputs, enc_valid_lens)  
            # å°†æ³¨æ„åŠ›ä¸Šä¸‹æ–‡å‘é‡ä¸å½“å‰æ—¶é—´æ­¥çš„è¾“å…¥è¿›è¡Œæ‹¼æ¥ï¼Œç”¨äºè¾“å…¥åˆ°å¾ªç¯ç¥ç»ç½‘ç»œ
            x = torch.cat((context, torch.unsqueeze(x, dim=1)), dim=-1)
            # æ‰§è¡Œå¾ªç¯ç¥ç»ç½‘ç»œçš„å‰å‘è®¡ç®—ï¼Œå¾—åˆ°è¾“å‡ºå’Œæ›´æ–°åçš„éšè—çŠ¶æ€
            out, hidden_state = self.rnn(x.permute(1,0,2),hidden_state)
            # å°†è¾“å‡ºå’Œæ³¨æ„åŠ›æƒé‡æ·»åŠ åˆ°å¯¹åº”çš„åˆ—è¡¨ä¸­
            outputs.append(out)
            self._attention_weights.append(self.attention.attention_weights)
        # å°†è¾“å‡ºåˆ—è¡¨ä¸­çš„ç»“æœè¿›è¡Œæ‹¼æ¥ï¼Œå¹¶é€šè¿‡çº¿æ€§å±‚è¿›è¡Œæ˜ å°„
        outputs = self.dense(torch.cat(outputs, dim=0))
        # å¯¹è¾“å‡ºè¿›è¡Œç»´åº¦å˜æ¢ï¼Œå°†batchç»´å’Œæ—¶é—´æ­¥ç»´äº¤æ¢ï¼Œå¹¶è¿”å›æ›´æ–°åçš„çŠ¶æ€ä¿¡æ¯
        return outputs.permute(1, 0, 2), [enc_outputs, hidden_state, enc_valid_lens]  
    
    @property
    def attention_weights(self):
        # è¿”å›æ³¨æ„åŠ›æƒé‡
        return self._attention_weights
```

```python
# æµ‹è¯•Bahdanauæ³¨æ„åŠ›è§£ç å™¨
# åˆ›å»ºç¼–ç å™¨å’Œè§£ç å™¨çš„å®ä¾‹
encoder = d2l.Seq2SeqEncoder(vocab_size = 10, embed_size=8, num_hiddens=16, num_layers=2)  
encoder.eval()
decoder = Seq2SeqAttentionDecoder(vocab_size=10, embed_size=8, num_hiddens=16,num_layers=2)  
decoder.eval()
# åˆ›å»ºè¾“å…¥åºåˆ—X
X = torch.zeros((4,7), dtype=torch.long)
# åˆå§‹åŒ–è§£ç å™¨çš„çŠ¶æ€
state = decoder.init_state(encoder(X), None)
# æ‰§è¡Œè§£ç å™¨çš„å‰å‘è®¡ç®—
output, state = decoder(X, state)
# è¾“å‡ºç»“æœçš„å½¢çŠ¶ä»¥åŠçŠ¶æ€ä¿¡æ¯çš„é•¿åº¦å’Œå½¢çŠ¶
output.shape, len(state), state[0].shape, len(state[1]), state[1][0].shape
```

```python
def read_data_nmt():
    """è½½å…¥ â€œè‹±è¯­-æ³•è¯­â€ æ•°æ®é›† """
    # ä¸‹è½½å¹¶è§£å‹æ•°æ®é›†
    data_dir = d2l.download_extract('fra-eng')
    # æ‰“å¼€æ•°æ®æ–‡ä»¶ï¼Œä½¿ç”¨utf-8ç¼–ç è¯»å–æ–‡ä»¶å†…å®¹
    with open(os.path.join(data_dir, 'fra.txt'), 'r', encoding='utf-8') as f:
        # è¿”å›è¯»å–çš„æ–‡ä»¶å†…å®¹
        return f.read()
```

```python
def preprocess_nmt(text):
    """é¢„å¤„ç† â€œè‹±è¯­-æ³•è¯­â€ æ•°æ®é›†"""
    def no_space(char, prev_char):
        # æ£€æŸ¥æ˜¯å¦éœ€è¦åœ¨å­—ç¬¦ä¹‹å‰æ·»åŠ ç©ºæ ¼
        return char in set(',.!?') and prev_char != ''
    # æ›¿æ¢ç‰¹æ®Šå­—ç¬¦ä¸ºç©ºæ ¼å¹¶è½¬æ¢ä¸ºå°å†™
    text = text.replace('\u202f', ' ').replace('\xa0',' ').lower()
    # æ ¹æ®è§„åˆ™åœ¨å­—ç¬¦ä¹‹å‰æ·»åŠ ç©ºæ ¼
    out = [
        ' ' + char if i > 0 and no_space(char, text[i - 1]) else char
        for i, char in enumerate(text)]
    # å°†å¤„ç†åçš„å­—ç¬¦åˆ—è¡¨è¿æ¥æˆå­—ç¬¦ä¸²å¹¶è¿”å›
    return ''.join(out)
```

```python
def tokenize_nmt(text, num_examples=None):
    """è¯å…ƒåŒ– â€œè‹±è¯­-æ³•è¯­â€ æ•°æ®æ•°æ®é›† """
    # åˆå§‹åŒ–æºè¯­è¨€å’Œç›®æ ‡è¯­è¨€çš„åˆ—è¡¨
    source, target = [], []
    # å¯¹æ¯è¡Œæ–‡æœ¬è¿›è¡Œéå†
    for i, line in enumerate(text.split('\n')):
        # å¦‚æœæŒ‡å®šäº†num_examplesï¼Œå¹¶ä¸”è¶…è¿‡äº†æŒ‡å®šæ•°é‡ï¼Œåˆ™é€€å‡ºå¾ªç¯
        if num_examples and i > num_examples:
            break
        # å°†æ¯è¡Œæ–‡æœ¬æŒ‰åˆ¶è¡¨ç¬¦åˆ†å‰²æˆæºè¯­è¨€å’Œç›®æ ‡è¯­è¨€çš„éƒ¨åˆ†
        parts = line.split('\t')
        # å¦‚æœåˆ†å‰²åæœ‰ä¸¤ä¸ªéƒ¨åˆ†ï¼Œè¯´æ˜åŒ…å«äº†æºè¯­è¨€å’Œç›®æ ‡è¯­è¨€çš„å†…å®¹
        if len(parts) == 2:
            # å°†æºè¯­è¨€éƒ¨åˆ†ä»¥ç©ºæ ¼ä¸ºåˆ†éš”ç¬¦è¿›è¡Œè¯å…ƒåŒ–ï¼Œå­˜å‚¨åˆ°æºè¯­è¨€åˆ—è¡¨ä¸­
            source.append(parts[0].split(' '))
            # å°†ç›®æ ‡è¯­è¨€éƒ¨åˆ†ä»¥ç©ºæ ¼ä¸ºåˆ†éš”ç¬¦è¿›è¡Œè¯å…ƒåŒ–ï¼Œå­˜å‚¨åˆ°ç›®æ ‡è¯­è¨€åˆ—è¡¨ä¸­
            target.append(parts[1].split(' '))
    # è¿”å›è¯å…ƒåŒ–åçš„æºè¯­è¨€åˆ—è¡¨å’Œç›®æ ‡è¯­è¨€åˆ—è¡¨
    return source, target
```

```python
def truncate_pad(line, num_steps, padding_token):
    """æˆªæ–­æˆ–å¡«å……æ–‡æœ¬åºåˆ—"""
    # å¦‚æœæ–‡æœ¬åºåˆ—çš„é•¿åº¦è¶…è¿‡äº†æŒ‡å®šçš„æ­¥æ•°
    if len(line) > num_steps:
        # æˆªæ–­æ–‡æœ¬åºåˆ—ï¼Œä¿ç•™å‰num_stepsä¸ªå…ƒç´ 
        return line[:num_steps]
    # åœ¨æ–‡æœ¬åºåˆ—æœ«å°¾å¡«å……padding_tokenï¼Œä½¿å…¶é•¿åº¦è¾¾åˆ°num_steps
    return line + [padding_token] * (num_steps - len(line))
```

```python
def build_array_nmt(lines, vocab, num_steps):
    """å°†æœºå™¨ç¿»è¯‘çš„æ–‡æœ¬åºåˆ—è½¬æ¢æˆå°æ‰¹é‡"""
    # å°†æ–‡æœ¬åºåˆ—ä¸­çš„æ¯ä¸ªè¯æ ¹æ®è¯æ±‡è¡¨è½¬æ¢ä¸ºå¯¹åº”çš„ç´¢å¼•
    lines = [vocab[l] for l in lines]
    # åœ¨æ¯ä¸ªæ–‡æœ¬åºåˆ—æœ«å°¾æ·»åŠ <eos>è¡¨ç¤ºå¥å­ç»“æŸçš„ç´¢å¼•
    lines = [l + [vocab['<eos>']] for l in lines]
    # å°†æ¯ä¸ªæ–‡æœ¬åºåˆ—æˆªæ–­æˆ–å¡«å……ä¸ºå›ºå®šé•¿åº¦ï¼Œå¹¶è½¬æ¢ä¸ºtensorå½¢å¼
    array = torch.tensor([ truncate_pad(l, num_steps, vocab['<pad>']) for l in lines ])
    # è®¡ç®—æœ‰æ•ˆé•¿åº¦ï¼Œå³éå¡«å……éƒ¨åˆ†çš„é•¿åº¦
    valid_len = (array != vocab['<pad>']).type(torch.int32).sum(1)
    # è¿”å›è½¬æ¢åçš„tensorå½¢å¼çš„æ–‡æœ¬åºåˆ—å’Œæœ‰æ•ˆé•¿åº¦
    return array, valid_len
```

```python
def load_data_nmt(batch_size, num_steps, num_examples=600):
    """è¿”å›ç¿»è¯‘æ•°æ®é›†çš„è¿­ä»£å™¨å’Œè¯æ±‡è¡¨"""
    # è½½å…¥å¹¶é¢„å¤„ç†æ–‡æœ¬æ•°æ®é›†
    text = preprocess_nmt(read_data_nmt())
    # å°†æ–‡æœ¬æ•°æ®é›†è¿›è¡Œè¯å…ƒåŒ–
    source, target = tokenize_nmt(text, num_examples)
    # æ„å»ºæºè¯­è¨€çš„è¯æ±‡è¡¨å¯¹è±¡
    src_vocab = d2l.Vocab(source, min_freq=2,
                         reserved_tokens=['<pad>','<bos>','<eos>'])
    # æ„å»ºç›®æ ‡è¯­è¨€çš„è¯æ±‡è¡¨å¯¹è±¡
    tgt_vocab = d2l.Vocab(target, min_freq=2,
                         reserved_tokens=['<pad>','<bos>','<eos>'])
    # å°†æºè¯­è¨€æ–‡æœ¬åºåˆ—è½¬æ¢ä¸ºå°æ‰¹é‡æ•°æ®ï¼Œå¹¶è·å–æœ‰æ•ˆé•¿åº¦
    src_array, src_valid_len = build_array_nmt(source, src_vocab, num_steps)  
    # å°†ç›®æ ‡è¯­è¨€æ–‡æœ¬åºåˆ—è½¬æ¢ä¸ºå°æ‰¹é‡æ•°æ®ï¼Œå¹¶è·å–æœ‰æ•ˆé•¿åº¦
    tgt_array, tgt_valid_len = build_array_nmt(target, tgt_vocab, num_steps)
    # æ„å»ºæ•°æ®é›†çš„æ•°ç»„
    data_arrays = (src_array, src_valid_len, tgt_array, tgt_valid_len)
    # æ„å»ºæ•°æ®è¿­ä»£å™¨
    data_iter = d2l.load_array(data_arrays, batch_size)
    # è¿”å›æ•°æ®è¿­ä»£å™¨å’Œè¯æ±‡è¡¨å¯¹è±¡
    return data_iter, src_vocab, tgt_vocab
```

```python
# è®­ç»ƒ
# è®¾ç½®åµŒå…¥å¤§å°ã€éšè—å±‚å¤§å°ã€å±‚æ•°å’Œä¸¢å¼ƒç‡
embed_size, num_hiddens, num_layers, dropout = 32, 32, 2, 0.1
# è®¾ç½®æ‰¹é‡å¤§å°å’Œåºåˆ—é•¿åº¦
batch_size, num_steps = 64, 10
# è®¾ç½®å­¦ä¹ ç‡ã€è®­ç»ƒè½®æ•°å’Œè®¾å¤‡ï¼ˆGPUæˆ–CPUï¼‰
lr, num_epochs, device = 0.005, 250, d2l.try_gpu()

# è½½å…¥ç¿»è¯‘æ•°æ®é›†ï¼Œå¹¶è·å–æ•°æ®è¿­ä»£å™¨å’Œè¯æ±‡è¡¨
train_iter, src_vocab, tgt_vocab = load_data_nmt(batch_size, num_steps)  
# åˆ›å»ºæºè¯­è¨€çš„ç¼–ç å™¨å®ä¾‹
encoder = d2l.Seq2SeqEncoder(len(src_vocab), embed_size, num_hiddens, num_layers, dropout)  
# åˆ›å»ºç›®æ ‡è¯­è¨€çš„å¸¦æœ‰æ³¨æ„åŠ›æœºåˆ¶çš„è§£ç å™¨å®ä¾‹
decoder = Seq2SeqAttentionDecoder(len(tgt_vocab), embed_size, num_hiddens, num_layers, dropout)   
# åˆ›å»ºç¼–ç -è§£ç æ¨¡å‹å®ä¾‹
net = d2l.EncoderDecoder(encoder, decoder)
# è®­ç»ƒåºåˆ—åˆ°åºåˆ—æ¨¡å‹
d2l.train_seq2seq(net, train_iter, lr, num_epochs, tgt_vocab, device)
```

```python
# å°†å‡ ä¸ªè‹±è¯­å¥å­ç¿»è¯‘æˆæ±‰è¯­
# è‹±è¯­å¥å­åˆ—è¡¨
engs = ['go', "i lost .", 'he\'s calm .', 'i\'m home .'] 
# å¯¹åº”çš„æ±‰è¯­å¥å­åˆ—è¡¨
fras = ['va !', 'j\' ai perdu .', 'il est calme .', 'je suis chez moi .']
# éå†è‹±è¯­å¥å­å’Œæ±‰è¯­å¥å­çš„å¯¹åº”å…³ç³»
for eng, fra in zip(engs, fras):
    # ä½¿ç”¨è®­ç»ƒå¥½çš„æ¨¡å‹netå¯¹è‹±è¯­å¥å­è¿›è¡Œç¿»è¯‘ï¼Œå¹¶è·å–æ³¨æ„åŠ›æƒé‡åºåˆ—
    translation, dec_attention_weight_seq = d2l.predict_seq2seq(net, eng, src_vocab,
                                                               tgt_vocab, num_steps, device, True)   
    # æ‰“å°è‹±è¯­å¥å­ã€ç¿»è¯‘ç»“æœå’ŒBLEUåˆ†æ•°
    print(f' {eng} => {translation}, ',
         f'bleu {d2l.bleu(translation, fra, k=2):.3f}')
```

```python
# å°†æ³¨æ„åŠ›æƒé‡åºåˆ—è¿›è¡Œæ‹¼æ¥ï¼Œå¹¶è°ƒæ•´å½¢çŠ¶
attention_weights = torch.cat([step[0][0][0] for step in dec_attention_weight_seq], 0).reshape((1, 1, -1, num_steps))     
```

```python
# å¯è§†åŒ–æ³¨æ„åŠ›æƒé‡
# æ˜¾ç¤ºæ³¨æ„åŠ›æƒé‡çš„çƒ­å›¾ï¼Œä»…æ˜¾ç¤ºä¸è¾“å…¥è‹±è¯­å¥å­å¯¹åº”çš„ä½ç½®
d2l.show_heatmaps(attention_weights[:,:,:,:len(engs[-1].split()) + 1].cpu(),
                 xlabel = 'key positions', ylabel = 'Query posistions')
```
