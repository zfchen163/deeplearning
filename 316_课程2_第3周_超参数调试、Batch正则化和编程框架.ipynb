# 5. TensorFlow

â‘  è¿™é‡Œç®€å•ä»‹ç»ä¸€ä¸‹æœ€è¿‘å‡ å¹´æ¯”è¾ƒç«çš„ä¸€ä¸ªæ·±åº¦å­¦ä¹ æ¡†æž¶ï¼šTensorFlowã€‚

â‘¡ ä¸¾ä¸ªä¾‹å­æ¥è¯´æ˜Žï¼Œä¾‹å¦‚cost functionæ˜¯å‚æ•°wçš„å‡½æ•°ï¼š

â‘¢ å¦‚æžœä½¿ç”¨TensorFlowå¯¹cost functionè¿›è¡Œä¼˜åŒ–ï¼Œæ±‚å‡ºæœ€å°å€¼å¯¹åº”çš„wï¼Œç¨‹åºå¦‚ä¸‹ï¼š

```python
import numpy as np
import tensorflow as tf
w = tf.Variable(0,dtype=tf.float32)
#cost = tf.add(tf.add(w**2,tf.multiply(-10,w)),25)
cost = w**2 - 10*w + 25
train = tf.train.GradientDescentOptimizer(0.01).minimize(cost)

init = tf.global_variables_initializer()
session = tf.Session()
session.run(init)

print(session.run(w))
```

```python
session.run(train)
print(session.run(w))
```

```python
for i in range(1000):
    session.run(train)
print(session.run(w))
```

â‘  TensorFlowæ¡†æž¶å†…å¯ä»¥ç›´æŽ¥è°ƒç”¨æ¢¯åº¦ä¸‹é™ä¼˜åŒ–ç®—æ³•ï¼Œä¸éœ€è¦æˆ‘ä»¬è‡ªå·±å†å†™ç¨‹åºäº†ï¼Œå¤§å¤§æé«˜äº†æ•ˆçŽ‡ã€‚åœ¨è¿è¡Œ1000æ¬¡æ¢¯åº¦ä¸‹é™ç®—æ³•åŽï¼Œwçš„è§£ä¸º4.99999ï¼Œå·²ç»éžå¸¸æŽ¥è¿‘wçš„æœ€ä¼˜å€¼5äº†ã€‚

â‘¡ é’ˆå¯¹ä¸Šé¢è¿™ä¸ªä¾‹å­ï¼Œå¦‚æžœå¯¹wå‰çš„ç³»æ•°ç”¨å˜é‡xæ¥ä»£æ›¿ï¼Œç¨‹åºå¦‚ä¸‹ï¼š

```python
import numpy as np
import tensorflow as tf

w = tf.Variable(0,dtype=tf.float32)
x = tf.placeholder(tf.float32,[3,1])
#cost = tf.add(tf.add(w**2,tf.multiply(-10,w)),25)
#cost = w**2 - 10*w +25
cost = x[0][0]*w**2 + x[1][0]*w + x[2][0]
train = tf.train.GradientDescentOptimizer(0.01).minimize(cost)

init = tf.global_variables_initializer()
session = tf.Session()
session.run(init)
print(session.run(w))
```

```python
coefficients = np.array([1,-10,25]).reshape(3, 1)
session.run(train, feed_dict={x: coefficients})
print(session.run(w))
```

```python
for i in range(1000):
    session.run(train, feed_dict={x: coefficients})
print(session.run(w))
```
