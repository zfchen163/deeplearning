## 🤔 为什么需要损失函数?

### 问题1: 不知道模型学得怎么样

**没有损失函数:**
```python
model.train()
# 训练了100轮
# 但是...学得怎么样? 不知道!
# 是好是坏? 不知道!
```

**有损失函数:**
```python
loss = loss_function(predictions, targets)
# loss = 0.5 → 学得不错
# loss = 10.0 → 学得很差
# 一目了然!
```

### 问题2: 无法指导训练

**没有损失函数:**
- 不知道参数应该怎么调整
- 不知道训练方向对不对
- 训练是盲目的

**有损失函数:**
- 损失大 → 需要调整参数
- 损失小 → 学得不错,继续
- 有明确的方向

### 问题3: 无法比较不同模型

**没有损失函数:**
- 模型A和模型B哪个更好?
- 不知道!
- 无法选择

**有损失函数:**
- 模型A损失=0.5
- 模型B损失=0.3
- 模型B更好! 一目了然

---


## 🎯 损失函数能解决什么问题?

### 1. **评估模型问题**
- ❌ 之前: 不知道模型学得怎么样
- ✅ 现在: 损失值直接反映模型好坏

### 2. **指导训练问题**
- ❌ 之前: 训练没有方向
- ✅ 现在: 损失函数指导参数调整

### 3. **模型比较问题**
- ❌ 之前: 无法比较不同模型
- ✅ 现在: 用损失值直接比较

### 4. **优化目标问题**
- ❌ 之前: 不知道优化什么
- ✅ 现在: 明确优化目标(最小化损失)

### 5. **训练停止问题**
- ❌ 之前: 不知道什么时候停止训练
- ✅ 现在: 损失不再下降就停止

## 📝 考试打分系统

想象你是老师,批改学生的试卷:

**学生的答案 vs 标准答案:**
- 完全正确 → 0分扣分(满分)
- 稍有偏差 → 扣5分
- 差距很大 → 扣20分
- 完全错误 → 扣50分

**扣分越少,说明学得越好!**

这就是损失函数的作用:
- 预测值 = 学生答案
- 真实值 = 标准答案
- 损失 = 扣分

## 🎯 射箭比赛

射箭时:
- 射中靶心(10环) → 损失 = 0(完美!)
- 射中9环 → 损失 = 1(很接近)
- 射中5环 → 损失 = 5(还行)
- 脱靶 → 损失 = 10(太差了)

**AI训练就是不断练习,让"损失"越来越小!**

## 🍕 外卖送达时间预测

外卖平台预测送达时间:
- 预测30分钟,实际30分钟 → 损失 = 0(完美)
- 预测30分钟,实际35分钟 → 损失 = 5(还行)
- 预测30分钟,实际60分钟 → 损失 = 30(太差,用户要投诉!)

**损失越小,用户越满意!**

# 6. 数据集计算损失函数

````python
import torch
import torchvision
from torch import nn 
from torch.nn import Conv2d, MaxPool2d, Flatten, Linear, Sequential
from torch.utils.data import DataLoader
from torch.utils.tensorboard import SummaryWriter

dataset = torchvision.datasets.CIFAR10("./dataset",train=False,transform=torchvision.transforms.ToTensor(),download=True)       
dataloader = DataLoader(dataset, batch_size=64,drop_last=True)

class Tudui(nn.Module):
    def __init__(self):
        super(Tudui, self).__init__()        
        self.model1 = Sequential(
            Conv2d(3,32,5,padding=2),
            MaxPool2d(2),
            Conv2d(32,32,5,padding=2),
            MaxPool2d(2),
            Conv2d(32,64,5,padding=2),
            MaxPool2d(2),
            Flatten(),
            Linear(1024,64),
            Linear(64,10)
        )
        
    def forward(self, x):
        x = self.model1(x)
        return x
    
loss = nn.CrossEntropyLoss() # 交叉熵    
tudui = Tudui()
for data in dataloader:
    imgs, targets = data
    outputs = tudui(imgs)
    result_loss = loss(outputs, targets) # 计算实际输出与目标输出的差距
    print(result_loss)
````

# 7. 损失函数反向传播

① 反向传播通过梯度来更新参数，使得loss损失最小，如下图所示。

````python
import torch
import torchvision
from torch import nn 
from torch.nn import Conv2d, MaxPool2d, Flatten, Linear, Sequential
from torch.utils.data import DataLoader
from torch.utils.tensorboard import SummaryWriter

dataset = torchvision.datasets.CIFAR10("./dataset",train=False,transform=torchvision.transforms.ToTensor(),download=True)       
dataloader = DataLoader(dataset, batch_size=64,drop_last=True)

class Tudui(nn.Module):
    def __init__(self):
        super(Tudui, self).__init__()        
        self.model1 = Sequential(
            Conv2d(3,32,5,padding=2),
            MaxPool2d(2),
            Conv2d(32,32,5,padding=2),
            MaxPool2d(2),
            Conv2d(32,64,5,padding=2),
            MaxPool2d(2),
            Flatten(),
            Linear(1024,64),
            Linear(64,10)
        )
        
    def forward(self, x):
        x = self.model1(x)
        return x
    
loss = nn.CrossEntropyLoss() # 交叉熵    
tudui = Tudui()
for data in dataloader:
    imgs, targets = data
    outputs = tudui(imgs)
    result_loss = loss(outputs, targets) # 计算实际输出与目标输出的差距
    result_loss.backward()  # 计算出来的 loss 值有 backward 方法属性，反向传播来计算每个节点的更新的参数。这里查看网络的属性 grad 梯度属性刚开始没有，反向传播计算出来后才有，后面优化器会利用梯度优化网络参数。      
    print("ok")
````
