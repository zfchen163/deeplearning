1. **ÂÖàÁêÜËß£"‰∏∫‰ªÄ‰πà"** - Ëøô‰∏™ÊäÄÊúØËß£ÂÜ≥‰ªÄ‰πàÂÆûÈôÖÈóÆÈ¢ò?

3. **ÊúÄÂêéÊéåÊè°"ÊÄé‰πàÂÅö"** - Â¶Ç‰ΩïÁî®‰ª£Á†ÅÂÆûÁé∞?
4. **Âä®ÊâãÂÆûË∑µ** - ËøêË°å‰ª£Á†Å,‰øÆÊîπÂèÇÊï∞,ËßÇÂØüÁªìÊûú

# 1. Â∫èÂàóÊ®°Âûã

# 2. Â∫èÂàóÊ®°Âûã

‚ë† ‰ΩøÁî®Ê≠£Âº¶ÂáΩÊï∞Âíå‰∏Ä‰∫õÂèØÂä†ÊÄßÂô™Â£∞Êù•ÁîüÊàêÂ∫èÂàóÊï∞ÊçÆÔºåÊó∂Èó¥Ê≠•‰∏∫1,2,...,1000

````python
%matplotlib inline
import torch
from torch import nn
from d2l import torch as d2l

# ÊÄªÊó∂Èó¥Ê≠•Êï∞
T = 1000 
# ÂàõÂª∫‰∏Ä‰∏™‰ªé1Âà∞TÁöÑÂº†Èáè
time = torch.arange(1, T + 1, dtype = torch.float32) # 1Âà∞1000‰∏∫Êó∂Èó¥
# ÁîüÊàêÈ¢ëÁéá‰∏∫0.01ÁöÑÊ≠£Âº¶Ê≥¢Âπ∂Âä†ÂÖ•ÈöèÊú∫Âô™Â£∞
x = torch.sin(0.01 * time) + torch.normal(0, 0.2, (T,))
# ÁªòÂà∂Êï∞ÊçÆ
d2l.plot(time, [x], 'time', 'x', xlim=[1,1000], figsize=(6,3))
````

‚ë° Â∞ÜÊï∞ÊçÆÊò†Â∞Ñ‰∏∫Êï∞ÊçÆÂØπ$y_t = x_t$Âíå$\mathbf{x}_t = [x_{t-\tau}, \ldots, x_{t-1}]$

````python
# Âª∂ËøüÊó∂Èó¥Ê≠•Èïø
tau = 4
# ÂàõÂª∫‰∏Ä‰∏™ÂΩ¢Áä∂‰∏∫(T - tau, tau)ÁöÑÈõ∂Âº†Èáè‰Ωú‰∏∫ÁâπÂæÅ
features = torch.zeros((T - tau, tau)) # T - tau ‰∏∫Ê†∑Êú¨Êï∞Ôºåtau ‰∏∫ÁâπÂæÅÊï∞
# ÊØèÂõõ‰∏™Êï∞ÊçÆ‰Ωú‰∏∫ÁâπÂæÅÔºåÁ¨¨‰∫î‰∏™‰Ωú‰∏∫Ê†áÁ≠æÔºå‰∏çÊñ≠ÊûÑÈÄ†ËøôÊ†∑ÁöÑÊï∞ÊçÆÂΩ¢ÊàêÊï∞ÊçÆÈõÜ
for i in range(tau):
    # ÊØèÂõõ‰∏™Êï∞ÊçÆ‰Ωú‰∏∫ÁâπÂæÅÔºåÁ¨¨‰∫î‰∏™‰Ωú‰∏∫Ê†áÁ≠æÔºå‰∏çÊñ≠ÊûÑÈÄ†ËøôÊ†∑ÁöÑÊï∞ÊçÆÂΩ¢ÊàêÊï∞ÊçÆÈõÜ
    features[:, i] = x[i:T - tau + i]
# ÊâÄ‰ªéÁ¨¨5‰∏™Êó∂ÂàªÂºÄÂßãÔºåÊØè‰∏™Êó∂ÂàªÁöÑlabelÊòØËØ•Êó∂ÂàªÁöÑxÂÄºÔºåËØ•Êó∂ÂàªÁöÑËæìÂÖ•ÊòØÂâç4‰∏™Êó∂ÂàªÁöÑÊï∞ÂÄºÁªÑÊàêÁöÑ‰∏Ä‰∏™ÂêëÈáè„ÄÇ
# ÁªèËøáÂèòÂåñÂêéÊï∞ÊçÆÁöÑËæìÂÖ•ÂÖ±Êúâ996ÁªÑ4‰∏™‰∏ÄÁªÑÁöÑÊï∞ÊçÆÔºåËæìÂá∫ÂÖ±996‰∏™ÂÄº
# ÊèêÂèñÊ†áÁ≠æÊï∞ÊçÆÂπ∂ËøõË°åÂΩ¢Áä∂ÂèòÊç¢
labels = x[tau:].reshape((-1,1))
# ÊâπÈáèÂ§ßÂ∞èÂíåËÆ≠ÁªÉÊ†∑Êú¨Êï∞Èáè
batch_size, n_train = 16, 600
# ‰ΩøÁî® features Âíå labels ÁöÑÂâç n_train ‰∏™Ê†∑Êú¨ÂàõÂª∫‰∏Ä‰∏™ÂèØËø≠‰ª£ÁöÑËÆ≠ÁªÉÈõÜ
train_iter = d2l.load_array((features[:n_train],labels[:n_train]),
                           batch_size, is_train=True)
````

## üîç Ê∑±ÂÖ•ÁêÜËß£: nn.ReLU

### ü§î ‰∏∫‰ªÄ‰πàË¶ÅÁî® ReLU?

**ÈóÆÈ¢ò: Á•ûÁªèÁΩëÁªúÂ§™"Á∫øÊÄß"‰∫Ü**
- Ê≤°ÊúâÊøÄÊ¥ªÂáΩÊï∞,Á•ûÁªèÁΩëÁªúÂè™ËÉΩÁîªÁõ¥Á∫ø
- Áé∞ÂÆû‰∏ñÁïåÂæàÂ§çÊùÇ,‰∏çÊòØÁõ¥Á∫øËÉΩËß£ÂÜ≥ÁöÑ

**Â∞±ÂÉè:**
- ‚ùå Ê≤°ÊúâReLU = Âè™ËÉΩÁî®Áõ¥Â∞∫ÁîªÂõæ
- ‚úÖ ÊúâReLU = ÂèØ‰ª•ÁîªÊõ≤Á∫ø„ÄÅÂúÜÂúà„ÄÅÂêÑÁßçÂΩ¢Áä∂

**‰∏çÁî®‰ºöÊÄéÊ†∑?**
‚ùå Â§öÂ±ÇÁ•ûÁªèÁΩëÁªúÁ≠â‰∫é‰∏ÄÂ±Ç(ÈÄÄÂåñ)
‚ùå Êó†Ê≥ïÂ≠¶‰π†Â§çÊùÇÊ®°Âºè
‚ùå ÂáÜÁ°ÆÁéáÂæà‰Ωé



### üì± ÂÆûÈôÖÂ∫îÁî®: Âà§Êñ≠Â§©Ê∞î

**Âú∫ÊôØ: Ê†πÊçÆÊ∏©Â∫¶„ÄÅÊπøÂ∫¶„ÄÅÈ£éÈÄüÂà§Êñ≠‰ºö‰∏ç‰ºö‰∏ãÈõ®**

```python
import torch
import torch.nn as nn

# ÂÅáËÆæÊàë‰ª¨Êúâ3‰∏™ËæìÂÖ•: Ê∏©Â∫¶„ÄÅÊπøÂ∫¶„ÄÅÈ£éÈÄü
# Ê∏©Â∫¶=25Â∫¶, ÊπøÂ∫¶=80%, È£éÈÄü=5m/s
weather_data = torch.tensor([[25.0, 80.0, 5.0]])

# Á¨¨1Ê≠•: Á∫øÊÄßÂ±Ç(ËÆ°ÁÆó)
linear = nn.Linear(3, 1)
score = linear(weather_data)
print(f"ËÆ°ÁÆóÂæóÂàÜ: {score.item():.2f}")  # ÂèØËÉΩÊòØË¥üÊï∞

# Á¨¨2Ê≠•: ReLUÊøÄÊ¥ª(ÂÜ≥Á≠ñ)
relu = nn.ReLU()
activated = relu(score)
print(f"ÊøÄÊ¥ªÂêé: {activated.item():.2f}")

# ReLUÁöÑ‰ΩúÁî®:
# - Â¶ÇÊûúÂæóÂàÜ < 0: ÂèòÊàê0(‰∏ç‰ºö‰∏ãÈõ®)
# - Â¶ÇÊûúÂæóÂàÜ > 0: ‰øùÊåÅ‰∏çÂèò(ÂèØËÉΩ‰∏ãÈõ®)

# ‰∏∫‰ªÄ‰πàËøôÊ†∑ÊúâÁî®?
```

### üìö Âæ™Â∫èÊ∏êËøõÁêÜËß£

**Á¨¨1Â±Ç: Â∞èÂ≠¶ÁîüÁêÜËß£**
- Ê≠£Êï∞‰∏çÂèò
- Ë¥üÊï∞ÂèòÊàê0

**Á¨¨2Â±Ç: Âàù‰∏≠ÁîüÁêÜËß£**
- ReLU = Rectified Linear Unit(‰øÆÊ≠£Á∫øÊÄßÂçïÂÖÉ)
- ÂÖ¨Âºè: f(x) = max(0, x)
- ‰ΩúÁî®: ËÆ©Á•ûÁªèÁΩëÁªúËÉΩÂ≠¶‰π†Â§çÊùÇÊ®°Âºè

**Á¨¨3Â±Ç: È´ò‰∏≠ÁîüÁêÜËß£**
- ‰∏∫‰ªÄ‰πàÈúÄË¶ÅÈùûÁ∫øÊÄß?
  * Á∫øÊÄßÂáΩÊï∞: y = ax + b(Áõ¥Á∫ø)
  * Â§öÂ±ÇÁ∫øÊÄß = ËøòÊòØÁ∫øÊÄß
  * Âä†ÂÖ•ReLU = ÂèØ‰ª•ÊãüÂêà‰ªªÊÑèÊõ≤Á∫ø

**Á¨¨4Â±Ç: ÂØπÊØîÂÖ∂‰ªñÊøÄÊ¥ªÂáΩÊï∞**
```python
x = torch.tensor([-2, -1, 0, 1, 2])

# ReLU: Ë¥üÊï∞Âèò0
relu = nn.ReLU()
print(f"ReLU: {relu(x)}")  # [0, 0, 0, 1, 2]

# Sigmoid: ÂéãÁº©Âà∞0-1
sigmoid = nn.Sigmoid()
print(f"Sigmoid: {sigmoid(x)}")  # [0.12, 0.27, 0.5, 0.73, 0.88]

# Tanh: ÂéãÁº©Âà∞-1Âà∞1
tanh = nn.Tanh()
print(f"Tanh: {tanh(x)}")  # [-0.96, -0.76, 0, 0.76, 0.96]
```

**‰∏∫‰ªÄ‰πàReLUÊúÄÂ∏∏Áî®?**
1. ËÆ°ÁÆóÁÆÄÂçï(ÈÄüÂ∫¶Âø´)
2. ‰∏ç‰ºöÊ¢ØÂ∫¶Ê∂àÂ§±(ËÆ≠ÁªÉÊïàÊûúÂ•Ω)
3. ÊïàÊûúÂ•Ω(Â§ßÂ§öÊï∞ÊÉÖÂÜµ)



````python
def init_weights(m):
    # Â¶ÇÊûúÂΩìÂâçÊ®°ÂùóÊòØÁ∫øÊÄßÂ±Ç
    if type(m) == nn.Linear:
        # ÂàùÂßãÂåñÊùÉÈáçÂáΩÊï∞ 
        nn.init.xavier_uniform_(m.weight)
    
def get_net():
    # ÂÆö‰πâÁ•ûÁªèÁΩëÁªúÁªìÊûÑ
    net = nn.Sequential(nn.Linear(4,10),nn.ReLU(),nn.Linear(10,1))
    # ÂØπÁΩëÁªúÁöÑÊùÉÈáçËøõË°åÂàùÂßãÂåñ
    net.apply(init_weights)
    # ËøîÂõûÊûÑÂª∫Â•ΩÁöÑÁ•ûÁªèÁΩëÁªúÊ®°Âûã
    return net

# ÂÆö‰πâÂùáÊñπËØØÂ∑ÆÊçüÂ§±ÂáΩÊï∞
loss = nn.MSELoss()
````

‚ë£ ËÆ≠ÁªÉÊ®°Âûã

````python
def train(net, train_iter, loss, epochs, lr):
    # ÂÆö‰πâ‰ºòÂåñÂô®
    trainer = torch.optim.Adam(net.parameters(), lr)
    # Ëø≠‰ª£ËÆ≠ÁªÉÊåáÂÆöÁöÑËΩÆÊï∞
    for epoch in range(epochs):
        # ÈÅçÂéÜËÆ≠ÁªÉÈõÜ‰∏≠ÁöÑÊØè‰∏™ÊâπÊ¨°
        for X, y in train_iter:
            # Ê¢ØÂ∫¶Ê∏ÖÈõ∂
            trainer.zero_grad()
            # ÂâçÂêë‰º†Êí≠ËÆ°ÁÆóÊçüÂ§±
            l = loss(net(X), y)
            # ÂèçÂêë‰º†Êí≠Ê±ÇÊ¢ØÂ∫¶
            l.backward()
            # Êõ¥Êñ∞Ê®°ÂûãÂèÇÊï∞
            trainer.step()
        # ÊâìÂç∞ÂΩìÂâçËΩÆÊ¨°ÁöÑÊçüÂ§±
        print(f'epoch {epoch + 1}, '
             f'loss: {d2l.evaluate_loss(net, train_iter, loss):f}')

# ÂàõÂª∫Á•ûÁªèÁΩëÁªúÊ®°Âûã
net = get_net()
# ËÆ≠ÁªÉÊ®°Âûã
train(net, train_iter, loss, 5, 0.01)
````

‚ë§ Ê®°ÂûãÈ¢ÑÊµã‰∏ã‰∏Ä‰∏™Êó∂Èó¥Ê≠•„ÄÇ

````python
# ÂØπÁâπÂæÅËøõË°å‰∏ÄÊ≠•È¢ÑÊµã
onestep_preds = net(features)
# ËøõË°åÊï∞ÊçÆÂèØËßÜÂåñÔºåÂ∞ÜÁúüÂÆûÊï∞ÊçÆÂíå‰∏ÄÊ≠•È¢ÑÊµãÁªìÊûúÁªòÂà∂Âú®Âêå‰∏Ä‰∏™Âõæ‰∏≠ËøõË°åÊØîËæÉ
d2l.plot([time, time[tau:]],
        [x.detach().numpy(), onestep_preds.detach().numpy()], 'time', 'x',
        legend = ['data','l-step preds'], xlim = [1, 1000], figsize=(6,3))   
````

‚ë• ËøõË°åÂ§öÊ≠•È¢ÑÊµã

````python
# ÂàùÂßãÂåñÂ§öÊ≠•È¢ÑÊµãÁªìÊûúÁöÑÂº†Èáè
multistep_preds = torch.zeros(T)
# Â∞ÜÂ∑≤Áü•ÁöÑÁúüÂÆûÊï∞ÊçÆËµãÂÄºÁªôÂ§öÊ≠•È¢ÑÊµãÁªìÊûú
multistep_preds[:n_train + tau] = x[:n_train + tau]
# ÂØπÂâ©‰ΩôÊó∂Èó¥Ê≠•ËøõË°åÂ§öÊ≠•È¢ÑÊµã
for i in range(n_train + tau, T):
    # Ëé∑ÂæóÂ§öÊ≠•È¢ÑÊµãÁªìÊûú
    multistep_preds[i] = net(multistep_preds[i - tau:i].reshape((1,-1)))

# ËøõË°åÊï∞ÊçÆÂèØËßÜÂåñ
d2l.plot(
    [time, time[tau:], time[n_train + tau:]], 
    [x.detach().numpy(), onestep_preds.detach().numpy(), multistep_preds[n_train + tau:].detach().numpy()],        
    'time', 
    'x', 
    legend = ['data', '1-step preds', 'multistep preds'], 
    xlim=[1,1000],
    figsize=(6,3) )  
````

‚ë¶ Êõ¥‰ªîÁªÜÂú∞Áúã‰∏Ä‰∏ãkÊ≠•È¢ÑÊµã„ÄÇ

````python
# ÊúÄÂ§ßÊ≠•Èïø
max_steps = 64
# ÂàùÂßãÂåñÁâπÂæÅÂº†Èáè
features = torch.zeros((T - tau - max_steps + 1, tau + max_steps))
# ‰ªé 0 Âà∞ tau-1ËøõË°åÈÅçÂéÜ
for i in range(tau):
    # ÊûÑÈÄ†ÁâπÂæÅÁü©Èòµ
    features[:, i] = x[i:i + T - tau - max_steps + 1]
    
# ‰ªé tau Âà∞ tau + max_steps - 1ÔºåÈÄöËøá net(features[:, i - tau:i]) ËøõË°åÂ§öÊ≠•È¢ÑÊµã
for i in range(tau, tau + max_steps):
    # ËøõË°åÂ§öÊ≠•È¢ÑÊµãÂπ∂Êõ¥Êñ∞ÁâπÂæÅÁü©Èòµ
    features[:,i] = net(features[:, i - tau:i]).reshape(-1)
    
# È¢ÑÊµãÁöÑÊ≠•Èïø
steps = (1, 4, 16, 64)
# ËøõË°åÊï∞ÊçÆÂèØËßÜÂåñ
d2l.plot([time[tau + i - 1:T - max_steps + i] for i in steps],
        [features[:, (tau + i - 1)].detach().numpy() for i in steps],
        'time',
        'x',
        legend = [f'{i}-step preds' for i in steps],
        xlim = [5,1000],
        figsize=(6,3) )
````
