```python
import hashlib
import os
import tarfile
import zipfile
import requests
import numpy as np
import pandas as pd
import torch
from torch import nn
from d2l import torch as d2l
%matplotlib inline

DATA_HUB = dict()
DATA_URL = 'http://d2l-data.s3-accelerate.amazonaws.com/'

def download(name, cache_dir=os.path.join('.', '01_data/02_DataSet_Kaggle_House')):
    """ä¸‹è½½ä¸€ä¸ªDATA_HUBä¸­çš„æ–‡ä»¶ï¼Œè¿”å›æœ¬åœ°æ–‡ä»¶å"""
    assert name in DATA_HUB, f"{name} ä¸å­˜åœ¨äº {DATA_HUB}."
    url , sha1_hash = DATA_HUB[name]
    os.makedirs(cache_dir, exist_ok=True)
    fname = os.path.join(cache_dir, url.split('/')[-1])
    if os.path.exists(fname):
        sha1 = hashlib.sha1()
        with open(fname,'rb') as f:
            while True:
                data = f.read(1048576)
                if not data:
                    break
                sha1.update(data)
        if sha1.hexdigest() == sha1_hash:
            return fname
    print(f'æ­£åœ¨ä»{url}ä¸‹è½½{fname}...')
    r = requests.get(url,stream=True,verify=True)
    with open(fname,'wb') as f:
        f.write(r.content)
    return fname
    
def download_extract(name, folder=None):
    """ä¸‹è½½å¹¶è§£å‹zip/taræ–‡ä»¶"""
    fname = download(name)
    base_dir = os.path.dirname(fname)
    data_dir, ext = os.path.splitext(fname)
    if ext == '.zip':
        fp = zipfile.ZipFile(fname, 'r')
    elif ext in ('.tar', '.gz'):
        fp = tarfile.open(fname, 'r')
    else:
        assert False, 'åªæœ‰zip/taræ–‡ä»¶å¯ä»¥è¢«è§£å‹ç¼©'
    fp,extractall(base_dir)
    return os.path.join(base_dir, folder) if folder else data_dir

def download_all():
    """ä¸‹è½½DATA_UHBä¸­çš„æ‰€æœ‰æ–‡ä»¶"""
    for name in DATA_HUB:
        download(name)

DATA_HUB['kaggle_house_train'] = (DATA_URL + 'kaggle_house_pred_train.csv','585e9cc9370b9160e7921475fbcd7d31219ce')         
DATA_HUB['kaggle_house_test'] = (DATA_URL + 'kaggle_house_pred_test.csv', 'fal9780a7b011d9b009e8bff8e99922a8ee2eb90')     
train_data = pd.read_csv(download('kaggle_house_train'))
test_data = pd.read_csv(download('kaggle_house_test'))
print(train_data.shape) # 1460ä¸ªæ ·æœ¬ï¼Œ80ä¸ªteç‰¹å¾ï¼Œ1ä¸ªæ ‡å·label
print(test_data.shape) # æµ‹è¯•æ ·æœ¬æ²¡æœ‰æ ‡å·label

print(train_data.iloc[0:4,[0,1,2,3,-3,-2,-1]]) # å‰é¢å››è¡Œçš„æŸäº›åˆ—ç‰¹å¾
```

```python
# å°†æ‰€æœ‰ç¼ºå¤±çš„å€¼æ›¿æ¢æˆç›¸åº”ç‰¹å¾çš„å¹³å‡å€¼
# é€šè¿‡å°†ç‰¹å¾é‡æ–°ç¼©æ”¾åˆ°é›¶å‡å€¼å’Œå•ä½æ–¹å·®æ¥æ ‡å‡†åŒ–æ•°æ®
print(all_features.dtypes) # å¯ä»¥çŸ¥é“æ¯ä¸€åˆ—åˆ†åˆ«ä¸ºä»€ä¹ˆç±»å‹ç‰¹å¾
numeric_features = all_features.dtypes[all_features.dtypes != 'object'].index  # å½“å€¼çš„ç±»å‹ä¸æ˜¯objectçš„è¯ï¼Œå°±æ˜¯ä¸€ä¸ªæ•°å€¼
print(numeric_features)
all_features[numeric_features] = all_features[numeric_features].apply(
    lambda x: (x - x.mean()) / (x.std())) # å¯¹æ•°å€¼æ•°æ®å˜ä¸ºæ€»ä½“ä¸ºå‡å€¼ä¸º0ï¼Œæ–¹å·®ä¸º1çš„åˆ†å¸ƒçš„æ•°æ®        
all_features[numeric_features] = all_features[numeric_features].fillna(0)  # å°†æ•°å€¼æ•°æ®ä¸­not numberçš„æ•°æ®ç”¨0å¡«å……      
```

```python
# å¤„ç†ç¦»æ•£å€¼ã€‚ç”¨ä¸€æ¬¡ç‹¬çƒ­ç¼–ç æ›¿æ¢å®ƒä»¬
# è‹¥ä¸€åˆ—é‡Œé¢æœ‰äº”ä¸ªä¸åŒçš„å€¼ï¼Œåˆ™åˆ›å»ºäº”ä¸ªfeaturesï¼Œå¦‚æœè¯¥åˆ—ä¸­ä¸ºè¯¥featureåˆ™ä¸º1ï¼Œä¸ä¸ºè¯¥featureåˆ™ä¸º0
all_features = pd.get_dummies(all_features,dummy_na=True) 
all_features.shape
```

```python
# è®­ç»ƒ
loss = nn.MSELoss()
print(train_features.shape[1]) # æ‰€æœ‰ç‰¹å¾ä¸ªæ•°
in_features = train_features.shape[1]
def get_net():
    net = nn.Sequential(nn.Linear(in_features,1)) # å•å±‚çº¿æ€§å›å½’
    return net
```

â‘  æˆ‘ä»¬æ›´å…³å¿ƒç›¸å¯¹è¯¯å·®$\frac{y - \hat{y}}{y}$ï¼Œè§£å†³è¿™ä¸ªé—®é¢˜çš„ä¸€ç§æ–¹æ³•æ˜¯ç”¨ä»·æ ¼é¢„æµ‹çš„å¯¹æ•°æ¥è¡¡é‡å·®å¼‚ã€‚

```python
def log_rmse(net, features, labels):
    clipped_preds = torch.clamp(net(features),1,float('inf')) # æŠŠæ¨¡å‹è¾“å‡ºçš„å€¼é™åˆ¶åœ¨1å’Œinfä¹‹é—´ï¼Œinfä»£è¡¨æ— ç©·å¤§ï¼ˆinfinityçš„ç¼©å†™ï¼‰       
    rmse = torch.sqrt(loss(torch.log(clipped_preds),torch.log(labels))) # é¢„æµ‹åšlogï¼Œlabelåšlogï¼Œç„¶åä¸¢åˆ°MSEæŸå¤±å‡½æ•°é‡Œ
    return rmse.item()
```

```python
# è®­ç»ƒå‡½æ•°å°†å€ŸåŠ©Adamä¼˜åŒ–å™¨
def train(net, train_features, train_labels, test_features, test_labels,
         num_epochs, learning_rate, weight_decay, batch_size):
    train_ls, test_ls = [], []
    train_iter = d2l.load_array((train_features, train_labels), batch_size)
    optimizer = torch.optim.Adam(net.parameters(), lr=learning_rate, weight_decay=weight_decay)
    for epoch in range(num_epochs):
        for X, y in train_iter:
            optimizer.zero_grad()
            l = loss(net(X),y)
            l.backward()
            optimizer.step()
        train_ls.append(log_rmse(net,train_features,train_labels))
        if test_labels is not None:
            test_ls.append(log_rmse(net, test_features, test_labels))
    return train_ls, test_ls
```

```python
# KæŠ˜äº¤å‰éªŒè¯
def get_k_fold_data(k,i,X,y): # ç»™å®škæŠ˜ï¼Œç»™å®šç¬¬å‡ æŠ˜ï¼Œè¿”å›ç›¸åº”çš„è®­ç»ƒé›†ã€æµ‹è¯•é›†
    assert k > 1
    fold_size = X.shape[0] // k  # æ¯ä¸€æŠ˜çš„å¤§å°ä¸ºæ ·æœ¬æ•°é™¤ä»¥k
    X_train, y_train = None, None
    for j in range(k): # æ¯ä¸€æŠ˜
        idx = slice(j * fold_size, (j+1)*fold_size) # æ¯ä¸€æŠ˜çš„åˆ‡ç‰‡ç´¢å¼•é—´éš”  
        X_part, y_part = X[idx,:], y[idx] # æŠŠæ¯ä¸€æŠ˜å¯¹åº”éƒ¨åˆ†å–å‡ºæ¥
        if j == i: # iè¡¨ç¤ºç¬¬å‡ æŠ˜ï¼ŒæŠŠå®ƒä½œä¸ºéªŒè¯é›†
            X_valid, y_valid = X_part, y_part
        elif X_train is None: # ç¬¬ä¸€æ¬¡çœ‹åˆ°X_trainï¼Œåˆ™æŠŠå®ƒå­˜èµ·æ¥ 
            X_train, y_train = X_part, y_part
        else: # åé¢å†çœ‹åˆ°ï¼Œé™¤äº†ç¬¬iå¤–ï¼Œå…¶ä½™æŠ˜ä¹Ÿä½œä¸ºè®­ç»ƒæ•°æ®é›†ï¼Œç”¨torch.catå°†åŸå…ˆçš„åˆå¹¶    
            X_train = torch.cat([X_train, X_part],0)
            y_train = torch.cat([y_train, y_part],0)
    return X_train, y_train, X_valid, y_valid # è¿”å›è®­ç»ƒé›†å’ŒéªŒè¯é›†
```

```python
# è¿”å›è®­ç»ƒå’ŒéªŒè¯è¯¯å·®çš„å¹³å‡å€¼
def k_fold(k, X_train, y_train, num_epochs, learning_rate, weight_decay,batch_size):
    train_l_sum, valid_l_sum = 0, 0
    for i in range(k):
        data = get_k_fold_data(k, i, X_train, y_train) # æŠŠç¬¬iæŠ˜å¯¹åº”åˆ†å¼€çš„æ•°æ®é›†ã€éªŒè¯é›†æ‹¿å‡ºæ¥   
        net = get_net()
        # *æ˜¯è§£ç ï¼Œå˜æˆå‰é¢è¿”å›çš„å››ä¸ªæ•°æ®
        train_ls, valid_ls = train(net, *data, num_epochs, learning_rate, weight_decay, batch_size) # è®­ç»ƒé›†ã€éªŒè¯é›†ä¸¢è¿›trainå‡½æ•° 
        
        train_l_sum += train_ls[-1]
        valid_l_sum += valid_ls[-1]
        
        if i == 0:
            d2l.plot(list(range(1, num_epochs + 1)), [train_ls,valid_ls],
                    xlabel='epoch',ylabel='rmse',xlim=[1,num_epochs],
                    legend=['train','valid'],yscale='log')
        print(f'fold{i+1},train log rmse {float(train_ls[-1]):f},'
             f'valid log rmse {float (valid_ls[-1]):f}')
    return  train_l_sum / k, valid_l_sum / k # æ±‚å’Œåšå¹³å‡
```

```python
# æ¨¡å‹é€‰æ‹©
k, num_epochs, lr, weight_decay, batch_size = 5, 100, 5, 0, 64
train_l, valid_l = k_fold(k, train_features, train_labels, num_epochs, lr, weight_decay, batch_size)   
print(f'{k}-æŠ˜éªŒè¯ï¼šå¹³å‡è®­ç»ƒlog rmseï¼š{float(train_l):f},'f'å¹³å‡éªŒè¯log rmseï¼š{float(valid_l):f}')    
```

```python
def train_and_pred(train_features, test_feature, train_labels, test_data, num_epochs, lr, weight_decay, batch_size):
    net = get_net()
    train_ls, _ = train(net, train_features, train_labels, None, None, num_epochs, lr, weight_decay, batch_size)  
    d2l.plot(np.arange(1, num_epochs + 1), [train_ls], xlabel='epoch',
            ylabel = 'log rmse', xlim=[1,num_epochs], yscale='log')
    print(f'train log rmse {float(train_ls[-1]):f}')
    preds = net(test_features).detach().numpy()
    test_data['SalePrice'] = pd.Series(preds.reshape(1,-1)[0])
    submission = pd.concat([test_data['Id'],test_data['SalePrice']],axis=1)
    submission.to_csv('submission.cvs',index = False)
    
train_and_pred(train_features, test_features, train_labels, test_data,
              num_epochs, lr, weight_decay, batch_size)
```
