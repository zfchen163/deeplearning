```python
import torch
from torch import nn
from d2l import torch as d2l

# ËÆæÁΩÆÊØè‰∏™batchÁöÑÂ§ßÂ∞èÂíåÊó∂Èó¥Ê≠•Êï∞
batch_size, num_steps = 32, 35
# Âä†ËΩΩÊó∂Èó¥Êú∫Âô®Êï∞ÊçÆÈõÜÂπ∂Ëé∑ÂèñËØçÊ±áË°®
train_iter, vocab = d2l.load_data_time_machine(batch_size, num_steps)
```

```python
# ÂàùÂßãÂåñÊ®°ÂûãÂèÇÊï∞
def get_params(vocab_size, num_hiddens, device):
    # ËÆæÁΩÆËæìÂÖ•ÂíåËæìÂá∫ÁöÑÁª¥Â∫¶‰∏∫ËØçÊ±áË°®Â§ßÂ∞è
    num_inputs = num_outputs = vocab_size
    
    # ÂÆö‰πâ‰∏Ä‰∏™ÂáΩÊï∞ÔºåÁî®‰∫éÁîüÊàêÊúç‰ªéÊ≠£ÊÄÅÂàÜÂ∏ÉÁöÑÈöèÊú∫Âº†ÈáèÔºåÂπ∂‰πò‰ª•0.01ËøõË°åÁº©Êîæ
    def normal(shape):
        return torch.randn(size=shape, device=device) * 0.01
    
    # ÂÆö‰πâ‰∏Ä‰∏™ÂáΩÊï∞ÔºåÁîüÊàê‰∏âÁªÑÊùÉÈáçÂíåÂÅèÁΩÆÂº†ÈáèÔºåÁî®‰∫é‰∏çÂêåÁöÑÈó®ÊéßÊú∫Âà∂
    def three():
        return (normal(
            (num_inputs, num_hiddens)), normal((num_hiddens, num_hiddens)),
                torch.zeros(num_hiddens, device=device))
    
    # ÂàùÂßãÂåñGRU‰∏≠ÁöÑÊùÉÈáçÂíåÂÅèÁΩÆ
    # ÊùÉÈáçÂíåÂÅèÁΩÆÁî®‰∫éÊéßÂà∂Êõ¥Êñ∞Èó®
    W_xz, W_hz, b_z = three() # GRUÂ§ö‰∫ÜËøô‰∏§Ë°å
    # ÊùÉÈáçÂíåÂÅèÁΩÆÁî®‰∫éÊéßÂà∂ÈáçÁΩÆÈó®
    W_xr, W_hr, b_r = three() # GRUÂ§ö‰∫ÜËøô‰∏§Ë°å
    # ÊùÉÈáçÂíåÂÅèÁΩÆÁî®‰∫éËÆ°ÁÆóÂÄôÈÄâÈöêËóèÁä∂ÊÄÅ
    W_xh, W_hh, b_h = three()
    # ÈöêËóèÁä∂ÊÄÅÂà∞ËæìÂá∫ÁöÑÊùÉÈáç
    W_hq = normal((num_hiddens, num_outputs))
    # ËæìÂá∫ÁöÑÂÅèÁΩÆ
    b_q = torch.zeros(num_outputs, device=device)
    # ÂèÇÊï∞ÂàóË°®‰∏≠ÂêÑÂèÇÊï∞È°∫Â∫è
    params = [W_xz, W_hz, b_z, W_xr, W_hr, b_r, W_xh, W_hh, b_h, W_hq, b_q]
    # ÈÅçÂéÜÂèÇÊï∞ÂàóË°®‰∏≠ÊâÄÊúâÂèÇÊï∞
    for param in params:
        # ËÆæÁΩÆÂèÇÊï∞ÁöÑ`requires_grad`Â±ûÊÄß‰∏∫TrueÔºå‰ª•‰æøËøõË°åÊ¢ØÂ∫¶ËÆ°ÁÆó
        param.requires_grad_(True)
    # ËøîÂõûÂèÇÊï∞ÂàóË°®‰∏≠ÊâÄÊúâÂèÇÊï∞
    return params
```

```python
# ÂÆö‰πâÈöêËóèÁä∂ÊÄÅÁöÑÂàùÂßãÂåñÂáΩÊï∞
def init_gru_state(batch_size, num_hiddens, device):
    # ËøîÂõûÈöêËóèÁä∂ÊÄÅÂàùÂßãÂåñ‰∏∫ÂÖ®Èõ∂ÁöÑÂÖÉÁªÑ
    return (torch.zeros((batch_size, num_hiddens), device=device),)
```

```python
# ÂÆö‰πâÈó®ÊéßÂæ™ÁéØÂçïÂÖÉÊ®°Âûã
def gru(inputs, state, params):
    # ÂèÇÊï∞ params Ëß£ÂåÖ‰∏∫Â§ö‰∏™ÂèòÈáèÔºåÂàÜÂà´Ë°®Á§∫Ê®°Âûã‰∏≠ÁöÑÊùÉÈáçÂíåÂÅèÁΩÆ
    W_xz, W_hz, b_z, W_xr, W_hr, b_r, W_xh, W_hh, b_h, W_hq, b_q = params     
    # ‰º†ÂÖ•ÁöÑÈöêËóèÁä∂ÊÄÅ state Ëß£ÂåÖ‰∏∫Âçï‰∏™ÂèòÈáè H„ÄÇ
    H, = state
    # ÂàõÂª∫‰∏Ä‰∏™Á©∫ÂàóË°®ÔºåÁî®‰∫éÂ≠òÂÇ®ÊØè‰∏™Êó∂Èó¥Ê≠•ÁöÑËæìÂá∫
    outputs = []
    # ÈÅçÂéÜËæìÂÖ•Â∫èÂàó‰∏≠ÁöÑÊØè‰∏™Êó∂Èó¥Ê≠•
    for X in inputs:
        # Êõ¥Êñ∞Èó®ÊéßÊú∫Âà∂ Z
        Z = torch.sigmoid((X @ W_xz) + (H @ W_hz) + b_z)
        # ÈáçÁΩÆÈó®ÊéßÊú∫Âà∂ R
        R = torch.sigmoid((X @ W_xr) + (H @ W_hr) + b_r)
        # ËÆ°ÁÆóÂÄôÈÄâÈöêËóèÁä∂ÊÄÅ H_tilda
        H_tilda = torch.tanh((X @ W_xh) + ((R * H) @ W_hh) + b_h)
        # Êõ¥Êñ∞ÈöêËóèÁä∂ÊÄÅ H
        H = Z * H + (1 - Z) * H_tilda
        # ËÆ°ÁÆóËæìÂá∫ Y
        Y = H @ W_hq + b_q
        # Â∞ÜËæìÂá∫Ê∑ªÂä†Âà∞ÂàóË°®‰∏≠
        outputs.append(Y)
    # Â∞ÜÊâÄÊúâËæìÂá∫ÊãºÊé•Âú®‰∏ÄËµ∑ÔºåÂπ∂ËøîÂõûÊãºÊé•ÂêéÁöÑÁªìÊûúÂíåÊúÄÁªàÁöÑÈöêËóèÁä∂ÊÄÅ
    return torch.cat(outputs, dim=0), (H,)
```

```python
# ËÆ≠ÁªÉ
vocab_size, num_hiddens, device = len(vocab), 256, d2l.try_gpu()
# ËÆæÁΩÆËÆ≠ÁªÉÁöÑÊÄªËΩÆÊï∞ÂíåÂ≠¶‰π†Áéá
num_epochs, lr = 500, 1
# ÂàõÂª∫ GRU Ê®°ÂûãÂÆû‰æã
model = d2l.RNNModelScratch(len(vocab), num_hiddens, device, get_params,
                           init_gru_state, gru)
# Ë∞ÉÁî® D2L Â∫ì‰∏≠ÁöÑËÆ≠ÁªÉÂáΩÊï∞Ôºå‰º†ÂÖ•Ê®°ÂûãÂÆû‰æã„ÄÅËÆ≠ÁªÉÊï∞ÊçÆËø≠‰ª£Âô®„ÄÅËØçÊ±áË°®„ÄÅÂ≠¶‰π†ÁéáÂíåËÆ≠ÁªÉÁöÑÊÄªËΩÆÊï∞
d2l.train_ch8(model, train_iter, vocab, lr, num_epochs, device)
```

```python
# ÁÆÄÊ¥ÅÂÆûÁé∞
# ËÆæÁΩÆËæìÂÖ•ÁâπÂæÅÁöÑÁª¥Â∫¶‰∏∫ËØçÊ±áË°®Â§ßÂ∞è
num_inputs = vocab_size
# ÂàõÂª∫‰∏Ä‰∏™ GRU Â±ÇÔºåÊåáÂÆöËæìÂÖ•ÁâπÂæÅÁöÑÁª¥Â∫¶ÂíåÈöêËóèÂçïÂÖÉÁöÑÊï∞Èáè
gru_layer = nn.GRU(num_inputs, num_hiddens)
# ÂàõÂª∫‰∏Ä‰∏™ RNNModel ÂÆû‰æãÔºå‰º†ÂÖ• GRU Â±ÇÂíåËØçÊ±áË°®ÁöÑÂ§ßÂ∞è
model = d2l.RNNModel(gru_layer, len(vocab))
# Â∞ÜÊ®°ÂûãÁßªÂä®Âà∞ÊåáÂÆöÁöÑËÆæÂ§á‰∏äÔºàÂèØËÉΩÊòØ GPUÔºâ
model = model.to(device)
# Ë∞ÉÁî® D2L Â∫ì‰∏≠ÁöÑ train_ch8 ÂáΩÊï∞ËøõË°åËÆ≠ÁªÉÔºå‰º†ÂÖ•Ê®°ÂûãÂÆû‰æã„ÄÅËÆ≠ÁªÉÊï∞ÊçÆËø≠‰ª£Âô®„ÄÅËØçÊ±áË°®„ÄÅÂ≠¶‰π†ÁéáÂíåËÆ≠ÁªÉÁöÑÊÄªËΩÆÊï∞
d2l.train_ch8(model, train_iter, vocab, lr, num_epochs, device)
```
