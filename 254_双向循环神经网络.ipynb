```python
import torch
from torch import nn
from d2l import torch as d2l

# ÂÆö‰πâÊâπÈáèÂ§ßÂ∞è„ÄÅÊó∂Èó¥Ê≠•Êï∞ÂíåËÆæÂ§áÔºàÂ¶ÇÊûúÂèØÁî®ÁöÑËØùÔºå‰ΩøÁî®GPUÔºâ
batch_size, num_steps, device = 32, 35, d2l.try_gpu()
# Âä†ËΩΩÊó∂Èó¥Êú∫Âô®Êï∞ÊçÆÈõÜÂπ∂ÁîüÊàêËÆ≠ÁªÉËø≠‰ª£Âô®ÂíåËØçÊ±áË°®
train_iter, vocab = d2l.load_data_time_machine(batch_size, num_steps)
# Ëé∑ÂèñËØçÊ±áË°®Â§ßÂ∞è„ÄÅÈöêËóèÂçïÂÖÉÊï∞ÈáèÂíåÂ±ÇÊï∞
vocab_size, num_hiddens, num_layers = len(vocab), 256, 2
# ËæìÂÖ•Â§ßÂ∞èÁ≠â‰∫éËØçÊ±áË°®Â§ßÂ∞è
num_inputs = vocab_size
# ÂàõÂª∫‰∏Ä‰∏™ÂèåÂêëLSTMÂ±Ç
lstm_layer = nn.LSTM(num_inputs, num_hiddens, num_layers, bidirectional=True)  # bidirectionalÊòØÂèåÂêëÂèÇÊï∞
# ‰ΩøÁî®ÂèåÂêëLSTMÂ±ÇÊûÑÂª∫RNNÊ®°Âûã
model = d2l.RNNModel(lstm_layer, len(vocab))
# Â∞ÜÊ®°ÂûãÁßªÂä®Âà∞ÊåáÂÆöËÆæÂ§áÔºàGPUÔºâ
model = model.to(device)
# ÂÆö‰πâËÆ≠ÁªÉÁöÑËΩÆÊï∞ÂíåÂ≠¶‰π†Áéá
num_epochs, lr = 500, 1
# ‰ΩøÁî®d2lÂ∫ì‰∏≠ÁöÑËÆ≠ÁªÉÂáΩÊï∞ËøõË°åÊ®°ÂûãËÆ≠ÁªÉÔºàÂèåÂêëÂæ™ÁéØÁ•ûÁªèÁΩëÁªúÊó†Ê≥ïÈ¢ÑÊµãÊú™Êù•‰ø°ÊÅØÔºâ
d2l.train_ch8(model, train_iter, vocab, lr, num_epochs, device) # ÂèåÂêëÂæ™ÁéØÁ•ûÁªèÁΩëÁªúÊó†Ê≥ïÈ¢ÑÊµãÊú™Êù•‰ø°ÊÅØ
```
